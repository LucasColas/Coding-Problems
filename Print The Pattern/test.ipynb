```
def tri_cocktail(board):
  echange = True 
  start = 0 
  end = len(board)-1

  while echange:

    echange = False

    for i in range(start,end):
      if board[i] > board[i+1]:
        board[i], board[i+1] = board[i+1], board[i]
        echange = True 

    if not echange:
      break

    echange = False

    end -= 1

    for i in range(end-1, start-1, -1):
      if board[i] > board[i+1]:
        board[i], board[i+1] = board[i+1], board[i]

        echange = True
    
    start += 1


tab = [4,5,1,3,9,12,11,17,6]

tri_cocktail(tab)
tab
```

```
def cocktailSort(a): 
    n = len(a) 
    swapped = True
    start = 0
    end = n-1
    while (swapped==True): 
  
        # reset the swapped flag on entering the loop, 
        # because it might be true from a previous 
        # iteration. 
        swapped = False
  
        # loop from left to right same as the bubble 
        # sort 
        for i in range (start, end): 
            if (a[i] > a[i+1]) : 
                a[i], a[i+1]= a[i+1], a[i] 
                swapped=True
  
        # if nothing moved, then array is sorted. 
        if (swapped==False): 
            break
  
        # otherwise, reset the swapped flag so that it 
        # can be used in the next stage 
        swapped = False
  
        # move the end point back by one, because 
        # item at the end is in its rightful spot 
        end = end-1
  
        # from right to left, doing the same 
        # comparison as in the previous stage 
        for i in range(end-1, start-1,-1): 
            if (a[i] > a[i+1]): 
                a[i], a[i+1] = a[i+1], a[i] 
                swapped = True
  
        # increase the starting point, because 
        # the last stage would have moved the next 
        # smallest number to its rightful spot. 
        start = start+1

Liste = [4,5,1,3,9,12,11,17,6]
cocktailSort(Liste)
Liste
```

```
def BubbleSort(board):
  for i in range(len(board)-1, 1, -1):
    for j in range(0, i):
      if board[j] > board[j+1]:
        temp = board[j]
        board[j] = board[j+1]
        board[j+1] = temp
  return board



BubbleSort([4,1,3,5,11,9,6])
```

```
def tri_fusion(Liste) : 
  if len(Liste) <= 1:
    return Liste 

  else:
    return fusion(tri_fusion(Liste[0:len(Liste)//2]), tri_fusion(Liste[len(Liste)//2:len(Liste)]))

def fusion(Liste1,Liste2):
  if len(Liste1) == 0:
    return Liste2

  if len(Liste2) == 0:
    return Liste1

  if Liste1[0] <= Liste2[0]:
    return [Liste1[0]] + fusion(Liste1[1:],Liste2)

  else:
    return [Liste2[0]] + fusion(Liste1,Liste2[1:])

Liste = [3,1,5,12,9,4,11]
tri_fusion(Liste)
```

```
def Mergesort(Board):
  if len(Board) <= 1:
    return Board

  else:
    #print(len(Board)//2)
    return Merge(Mergesort(Board[0:len(Board)//2]), Mergesort(Board[len(Board)//2:len(Board)]))


def Merge(Board_A,Board_B):
  #print("Board B" ,Board_B)
  #print("Board A ", Board_A)

  if len(Board_A) <= 0:
    return Board_B

  if len(Board_B) <= 0:
    return Board_A

  if Board_A[0] <= Board_B[0]:

    return [Board_A[0]] + Merge(Board_A[1:], Board_B)

  else:

    return [Board_B[0]] + Merge(Board_A, Board_B[1:])


Mergesort([3,1,5,12,9,4,11])
```

```
def prime_factors(num):
    output = []
    for i in range(2,  int(num/2)):
        while num % i == 0:
            output.append(i)
            num = int(num / i)

    return output

prime_factors(300)
```

```
from math import sqrt
def produit_fac_premiers(num):
  facteurs_premiers = []
  for n in range(2, num):
    while num % n == 0:
      facteurs_premiers.append(n)
      num //=  n
     

  return facteurs_premiers


produit_fac_premiers(236)
```

```
import math
def get_prime_factors(number):
    # create an empty list and later I will
    # run a for loop with range() function using the append() method to add elements to the list.
    prime_factors = []

    # First get the number of two's that divide number
    # i.e the number of 2's that are in the factors
    while number % 2 == 0:
        prime_factors.append(2)
        number = number / 2

    # After the above while loop, when number has been
    # divided by all the 2's - so the number must be odd at this point
    # Otherwise it would be perfectly divisible by 2 another time
    # so now that its odd I can skip 2 ( i = i + 2) for each increment
    for i in range(3, int(math.sqrt(number)) + 1, 2):
        while number % i == 0:
            prime_factors.append(int(i))
            number = number / i

    return prime_factors


get_prime_factors(28)
```

```
def max_ligne(matrice):
  max_matrice = [[0 for i in range(len(matrice[0]))] for j in range(len(matrice))]
  for i in range(len(matrice)):
    max = matrice[i][0]
    for index, valeur in enumerate(matrice[i]):
      if valeur > max:
        max = valeur 

    for j in range(len(matrice[i])):
      if matrice[i][j] == max:
        max_matrice[i][j] = 1 

  return max_matrice
```

```
def min_colonne(matrice):
  min_matrice = [[0 for i in range(len(matrice[0]))] for j in range(len(matrice))]

  for i in range(len(matrice[0])):
    min = matrice[0][i]
    for j in range(len(matrice)):
      if matrice[j][i] < min:
        min = matrice[j][i]

    for k in range(len(matrice)):
      if matrice[k][i] == min:
        min_matrice[k][i] = 1 

  return min_matrice
  
min_matrice = min_colonne(matrice)
afficher_matrice(min_matrice)
```

```
def verif_ligne(M, somme):
  for ligne in M:
    som = 0
    for chiffre in ligne:
      som += chiffre

    if som != somme:
      print("Une ligne n'est pas égale à la somme rentrée")
      return 

  print("La somme de chaque ligne est égale à la somme rentrée")

M = [[2,7,6],[9,5,1],[4,3,8]]
verif_ligne(M, 15)
```

```
def verif_colonne(M, somme):
  for col in range(len(M)):
    som = 0
    for row in M:
      som += row[col]

    if som != somme:
      print("Une ligne n'est pas égale à la somme rentrée")
      return 

  print("La somme de chaque ligne est égale à la somme rentrée")


M = [[2,7,6],[9,5,1],[4,3,8]]
verif_colonne(M, 15)

    
```

```
def verif_diagonale1(M, somme):
  som = 0
  for index in range(len(M)):
    som += M[index][index]

  if som != somme:
    print("Une diagonale n'est pas égale à la somme rentrée")
    return 

  print("La somme de la diagonale est égale à la somme rentrée")


M = [[2,7,6],[9,5,1],[4,3,8]]
verif_diagonale1(M, 15)
```

```
def verif_diagonale2(M, somme):
  som = 0
  for index, rev_index in enumerate(reversed(range(len(M)))):
    som += M[index][rev_index]

  if som != somme:
    print("Une diagonale n'est pas égale à la somme rentrée")
    return 

  print("La somme de la deuxième diagonale est égale à la somme rentrée")


M = [[2,6,7],[9,5,1],[4,3,8]]
verif_diagonale2(M,15)
```
